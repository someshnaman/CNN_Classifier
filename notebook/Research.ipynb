```python
TARGET_DATA_DIRS = ["Cat", "Dog"]
PARENT_DIR = os.path.join("data","Petimages")
```

```python
from PIL import Image
```

```python
BAD_DATA_DIR = "bad_data"
os.makedirs(BAD_DATA_DIR, exist_ok=True)
```

```python
import shutil
import imghdr
```

```python
for dirs  in os.listdir(PARENT_DIR):
    full_path = os.path.join(PARENT_DIR,dirs)
    for imgs in os.listdir(full_path):        
        path_tp_imgs = os.path.join(full_path, imgs)
        print(path_tp_imgs)
        try:
            imgs = Image.open(path_tp_imgs)
            imgs.verify()
            print(f'{path_tp_imgs} is verified')
        except Exception as e:
            print(f'{path_tp_imgs} is bad')
            bad_data_path=os.path.join(BAD_DATA_DIR,imgs)
            shutil.move(path_tp_imgs, bad_data_path)
        
    
```

```python
IMG_SIZE = (180,180)
BATCH_SIZE = 32
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    PARENT_DIR,
    validation_split=.2,
    subset = 'training',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    seed=42
)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    PARENT_DIR,
    validation_split=.2,
    subset = 'validation',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    seed=42
)
```

```python
import time
import numpy as np
def get_log_path(base_log_dir=os.path.join("logs","fit")):
    uniqueName = time.asctime().replace(" ", "_").replace(":", "")
    log_path = os.path.join(base_log_dir, uniqueName)
    print(f"saving logs at: {log_path}")
    return log_path

log_dir = get_log_path()
```

```python
file_writer = tf.summary.create_file_writer(logdir=log_dir)
```

```python
for images, labels in train_ds.take(1):
    print(images.shape, labels)
```

```python
with file_writer.as_default():
    images = np.array(images) ### (20, 28, 28, 1)
    tf.summary.image("samples", images.astype("uint8"), max_outputs=25, step=0,)
  
```
